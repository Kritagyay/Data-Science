## Histograms and Subplots

```python
# make a histogram plot 
x=np.random.randn(1000)

fig,ax=plt.subplots()
ax.hist(x)
plt.show()
```

```python
# make a subplot 
#Option 1 

fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,8))

ax1.plot(x,x/2)
ax2.scatter(np.random.randn(100),np.random.randn(100))
ax3.bar(list(new_dict.keys()),list(new_dict.values()))
ax4.hist(x)
plt.show()
```

```python
#option2 for the subplots 
fig,ax=plt.subplots(figsize=(10,8),
                   nrows=2,
                   ncols=2)
ax[0][0].plot(x,x/2)
ax[0][1].scatter(np.random.randn(100),np.random.randn(100))
ax[1][0].hist(np.random.randn(150))
ax[1][1].bar(list(new_dict.keys()),list(new_dict.values()))
plt.show()
```

##  Two Methods to solve the question
* First is ther pyplot method
* Second method is OO method which is used for the complex problems

```python
heart_diss.head()
```

```python
over_age=heart_diss[heart_diss["age"]>50]
over_age
```

```python
over_age.plot.scatter(figsize=(10,6),
                       x="age",
                       y="chol",
                       c="target");
```

```python
# Now we will opt for the OO method 
fig,ax=plt.subplots(figsize=(10,5))
over_age.plot(kind="scatter",
             x="age",
             y="chol",
              c="target",
             ax=ax);
```

```python
# Now we will only perform with the OO method
over_age
```

```python
fig,ax=plt.subplots(figsize=(10,5))
scatter=ax.scatter(over_age["age"],over_age["chol"],
                  c=over_age["target"]);
ax.set(title="Heart Dissease to people over 50 years due to Cholestrol",
        xlabel="Age",
        ylabel="Cholestrol Level");

ax.legend(*scatter.legend_elements(),title="Target");
ax.axhline(over_age["chol"].mean(),color="b",linestyle="--")
```

```python
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

scatter=ax0.scatter(over_age["age"],
                    over_age["chol"],
               c=over_age["target"])

ax0.set(title="Heart_Dissease due to cholestrol",
       ylabel="Cholestrol")

ax0.legend(*scatter.legend_elements(),title="Target")

ax0.axhline(over_age["chol"].mean(),color="b",linestyle="--")


scatter=ax1.scatter(over_age["age"],over_age["thalach"],c=over_age["target"])

ax1.set(title="Heart_Dissease due to Thalach",
       xlabel="Age",
       ylabel="Max Heart Rate")

ax1.legend(*scatter.legend_elements(),title="Target")

ax1.axhline(over_age["thalach"].mean(),linestyle="--",color="b")
fig.suptitle("Heart Dissese Chart",fontsize=15,fontweight="bold")
```
